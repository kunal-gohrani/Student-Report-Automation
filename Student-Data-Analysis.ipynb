#### Spelling error fixings
From the above 2 rows, we can see that there are different names in the data, Sushree and Siddhishikha is the same person, same for Swaastik and Swaastick.

```python
data_july.replace('Swaastik','Swaastick',inplace=True)
data_august.replace('Swaastik','Swaastick',inplace=True)
```

```python
data_july.replace('Sushree','Siddhishikha',inplace=True)
data_august.replace('Sushree','Siddhishikha',inplace=True)
```

```python
np.sort(data_august['Student'].unique())
```
