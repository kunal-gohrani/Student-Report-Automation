```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import imagekit
import imgkit
%matplotlib inline
```

```python
excel_sheet=pd.read_excel('Student Gradebook.xlsx',sheet_name=None,usecols=range(0,10))
```

```python
data_july=excel_sheet['July'].copy()
data_august=excel_sheet['August'].copy()
```

```python
data_july.head()
```

```python
data_july=data_july.dropna(how='all',axis=0)
```

```python
data_july.info()
```

```python
data_august.info()
```

# Student column

We know from above info(), data_august has some null values, let's see to it

```python
data_august[data_august['Student'].isnull()]
```

Since, there is no way to correctly fill this data, we are just gonna drop those rows.

```python
data_august.info()
```

```python
np.sort(data_july['Student'].unique())
```

```python
np.sort(data_august['Student'].unique())
```

#### Spelling error fixings
From the above 2 rows, we can see that there are different names in the data, Sushree and Siddhishikha is the same person, same for Swaastik and Swaastick.

```python
data_july.replace('Swaastik','Swaastick',inplace=True)
data_august.replace('Swaastik','Swaastick',inplace=True)
```

```python
data_july.replace('Sushree','Siddhishikha',inplace=True)
data_august.replace('Sushree','Siddhishikha',inplace=True)
```

```python
np.sort(data_july['Student'].unique())
```

```python
np.sort(data_august['Student'].unique())
```

The name problem has been resolved.

```python
data_august
```

```python
marksaug=data_august.groupby('Student')['Points'].sum().sort_values(ascending=False)
marksjuly=data_july.groupby('Student')['Points'].sum().sort_values(ascending=False)
total_marks=(marksjuly+marksaug).sort_values(ascending=False).reset_index().dropna()
```

```python
total_marks
```
